```python
from hpacker import HPacker

# model_dirs is simply two-element tuple/list of directories containing the "initial_guess" and "refinement" models
model_dirs = [
    'pretrained_models/initial_guess',
    'pretrained_models/refinement',
]
```

```python
## Reconstruct ground truth structure, evaluate difference

pdbpath = '/gscratch/scrubbed/gvisan01/dlpacker/casp_targets/T0950.pdb'

hpacker = HPacker(model_dirs, pdbpath, remove_sidechains=True)

metrics = hpacker.reconstruct_sidechains_and_evaluate(num_refinement_iterations = 2, batch_size = 128, standardize_original_structure = False)

mae_per_angle_4, accuracy_per_angle_4, real_chis, predicted_chis, aas, res_ids_dict, rmsds = metrics



```
